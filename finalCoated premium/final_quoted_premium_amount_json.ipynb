```python
in_data = row_data['customer_request']
```

```python
for key in in_data:
    temp = in_data[key].split('+')
    temp[0] = int(temp[0])
    temp[1] = int(temp[1])
    temp[3] = int(temp[3])
    in_data[key] = temp
```

```python
def Dwelling_cover(dc):
    
    xs = np.array([100000,150000,200000,250000,300000,350000])
    ys = np.array([0.971,1.104,1.314,1.471,1.579,1.761]) 
    m,b = np.polyfit(xs,ys,1)
    
    y_new = m*dc+b    
    return y_new
```

```python
def Home_age(ag):
    if ag >= 0 and ag <= 10:
        return 1.00
    elif ag >= 11 and ag <= 35:
        return 1.50
    elif ag >= 36 and ag <= 100:
        return 1.80
    elif ag > 100:
        return 1.95
    else:
        raise Exception()
```

```python
def Roof_type(r_type):
    if r_type == 'Asphalt Shingles':
        return 1.00
    elif r_type == 'Tin':
        return 1.70
    elif r_type == 'Wood':
        return 2.00
    else:
        raise Exception()
```

```python
def Number_unit(un):
    if un == 1:
        return 1.00
    if un == 2 or un == 3 or un == 4:
        return 0.80

    else:
        raise Exception()    
```

```python
out_data = {}
```

```python
def discount_rate(idd):
    if idd == 'Y':
        return 0.05
    else:
        return 0
```

```python
for key in in_data:
    dcc,ha,rt,nu,idd = in_data[key]
    print(dcc, nu, ha, rt, idd)
    
    base_premium = 350
    
   
    
    primium_amount = base_premium*Dwelling_cover(dcc)*Home_age(ha)*Roof_type(rt)*Number_unit(nu)
    
    discount = primium_amount*discount_rate(idd)
    
    final_quoted_premium_amount = primium_amount - discount
    
    out_data[key] = round(final_quoted_premium_amount,0)

    
print(out_data)
    
```

```python
json_output = {'quoted_premium' : out_data}
fo = open('output','w')
fo.write(json.dumps(json_output))
fo.close()
```
