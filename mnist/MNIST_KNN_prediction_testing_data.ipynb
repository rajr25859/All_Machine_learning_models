```python
from sklearn.neighbors import KNeighborsClassifier
from matplotlib import pyplot as plt
import numpy as np
from mnist import MNIST
import png
import os
import pickle
from sklearn.metrics import classification_report, accuracy_score
```

```python
mnist_data = MNIST(os.path.join(os.getcwd(), 'D:\DATA FOR MACHINE LEARNING PROJECT\MNIST_model_project\mlengine\mnist_data'))
X_train, y_train = mnist_data.load_training()
X_test , y_test = mnist_data.load_testing()
```

```python
X_train = np.asarray(X_train)
y_train = np.asarray(y_train)
X_test = np.asarray(X_test)
y_test = np.asarray(y_test)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
pred_knn = knn.predict(X_test)
```

```python
accuracy_score(y_test,pred_knn)
```

```python
cr = classification_report(y_test,pred_knn)
print(cr)
```
