```python
import numpy as np
import pandas as pd
from mnist import MNIST
```

```python
df = MNIST('C:\\Users\\acer\Downloads\\New folder')
```

```python
X_train, y_train = df.load_training()
X_test, y_test = df.load_testing()
```

```python
X_train = np.asarray(X_train)
y_train = np.asarray(y_train) 
```

```python
X_test = np.asarray(X_test)
y_test = np.asarray(y_test)
```

```python
X_test.shape, y_train.shape,
```

```python
X_train.shape, y_train.shape
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

---

### Random Forest Classifier

```python
rfc = RandomForestClassifier()
rfc = rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
classification_report(y_test, y_pred)
```

---

### KNeighborsClassifier

```python
knn = KNeighborsClassifier()
knn = knn.fit(X_train, y_train)
```

```python
y_pred2 = knn.predict(X_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
classification_report(y_test, y_pred2)
```

---

### DecisionTreeClassifier

```python
dtc = DecisionTreeClassifier()
dtc = dtc.fit(X_train, y_train)
```

```python
y_pred3 = dtc.predict(X_test)
```

```python
accuracy_score(y_test,y_pred3)
```

```python
classification_report(y_test, y_pred3)
```

---

### GaussianNB

```python
gn = GaussianNB()
gn = gn.fit(X_train, y_train)
```

```python
y_pred4 = gn.predict(X_test)
```

```python
accuracy_score(y_test,y_pred4)
```

```python
classification_report(y_test, y_pred4)
```

---

### LogisticRegression

```python
lgr = LogisticRegression()
lgr = lgr.fit(X_train, y_train)
```

```python
y_pred5 = lgr.predict(X_test)
```

```python
accuracy_score(y_test,y_pred5)
```

```python
classification_report(y_test, y_pred5)
```

### SVC

```python
sv = SVC(gamma='auto')
sv = sv.fit(X_train, y_train)
```
