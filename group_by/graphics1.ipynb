```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
plt.rcParams['figure.figsize'] = 18,10
```

```python
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset('tips')
```

```python
print(tips.shape)
```

```python
print(tips)
```

```python
tips.columns
```

```python
tips.columns = ['total_bill','tip','gender','smoker','day','time','size']
```

```python
tips.columns
```

```python
tips
```

## Relation between Tips and Total Bill

```python
sns.scatterplot(data = tips,x = 'total_bill', y = 'tip')
plt.show()
```

## Relation between Size and Total Bill

```python
sns.scatterplot(tips['size'],tips['total_bill'])
plt.show()
```

## Spread of Total Bill

```python
plt.subplot(1,2,1)
sns.stripplot(data = tips, y = 'total_bill')

plt.subplot(1,2,2)
sns.swarmplot(data = tips, y = 'total_bill')
plt.show()
```

## Time wise Total Bill Spread


```python
plt.subplot(1,2,1)
sns.stripplot(data = tips, x = 'time', y = 'total_bill')

plt.subplot(1,2,2)
sns.swarmplot(data = tips, x = 'time', y = 'total_bill')
plt.show()
```

## Time wise Total Bill with hue effects

```python
plt.subplot(1,2,1)
sns.stripplot(data = tips, x = 'time', y = 'total_bill',hue = 'day')

plt.subplot(1,2,2)
sns.swarmplot(data = tips, x = 'time', y = 'total_bill', hue = 'day')
plt.show()
```

## Time wise Total Bill with hue and dodge effects

```python
plt.subplot(1,2,1)
sns.stripplot(data = tips, x = 'time', y = 'total_bill',hue = 'day', dodge = True)

plt.subplot(1,2,2)
sns.swarmplot(data = tips, x = 'time', y = 'total_bill',hue = 'day', dodge = True)
plt.show()
```

## Time wise Total Amount with hue, dodge, pallete, marker, size, alpha

```python
plt.subplot(1,2,1)
sns.stripplot(data = tips, x = 'time', y = 'total_bill',hue = 'day', dodge = True,palette = 'bright',marker = 'D', size = 10,alpha = 0.5)

plt.subplot(1,2,2)
sns.swarmplot(data = tips, x = 'time', y = 'total_bill',hue = 'day', dodge = True,palette = 'bright',marker = 'D', size = 10,alpha = 0.5)
plt.show()
```

```python
result1 = tips.groupby(['time']).agg({
    'size': 'sum'})
result1.reset_index(inplace = True)
print(result1)
```

## Time wise Seats Booked

```python
plt.figure(figsize = (7,7))
plt.title('Seat Bookings as per Timing')
plt.pie(result1['size'],labels = result1['time'],autopct = '%.1f%%',startangle = 90,explode = [0,0.1])
plt.show()
```
