```python
data = df[['InvoiceDate','InvoiceNo','StockCode','UnitPrice','Quantity','Amount','CustomerID']]
```

```python
print(data.info())
```

# Invoice wise count of products, Number of Units Sold and Total Amount

```python
result1 = data.groupby(['InvoiceDate','CustomerID','InvoiceNo']).agg({
    'StockCode': 'count',
    'Quantity' : 'sum',
    'Amount'   : 'sum'})
result1.columns=['Products','Units','Total Amount']
print(result1)
```

## Date Wise Number of Invoice, Number of Units Sold, Total Amount

```python
result2 = data.groupby(['InvoiceDate','InvoiceNo']).agg({
    'Quantity' : 'sum',
    'Amount'   : 'sum' })
print(result2)
```

```python
result2.reset_index(inplace = True)
print(result2)
```

## InvoiceDate wise Invoices, Units, Total Amount

```python
result3 = data.groupby(['InvoiceDate']).agg({
    'InvoiceNo' : 'count',
    'Quantity'      : 'sum',
    'Amount'        : 'sum' })

result3.columns = ['Invoices','Units','Total Amount']
print(result3)
```

## Date wise Invoices, Units, Total Amount

```python
result4 = result3.groupby(pd.Grouper(freq = 'D')).sum()
result4.reset_index(inplace = True)
print(result4)
```

## Month wise Invoices, Units, Total Amount

```python
result5 = result3.groupby(pd.Grouper(freq = 'M')).sum()
result5.reset_index(inplace = True)
print(result5)
```

## Year wise Invoices, Units, Total Amount

```python
result6 = result3.groupby(pd.Grouper(freq = 'Y')).sum()
result6.reset_index(inplace = True)
print(result6)
```

## StockCode wise Amount

```python
result7 = data.groupby(['StockCode']).agg({'Amount':'sum'})
print(result7)
```

## CustomerID wise Invoices, Units, Total Amount

```python
result8 = data.groupby(['CustomerID']).agg({
    'InvoiceNo' : 'count',
    'Quantity'      : 'sum',
    'Amount'        : 'sum' })

result8.columns = ['Invoices','Units','Total Amount']
result8 = result8.reset_index()
print(result8)
```

## CustomerID wise Total Amount

```python
result9 = result8[['CustomerID','Total Amount']].sort_values('Total Amount',ascending = False).head(10)
print(result9)
```

## CustomerID wise Units

```python
result10 = result8[['CustomerID','Units']].sort_values('Units',ascending = False).head(10)
print(result10)
```

## CustomerID wise Invoices

```python
result11 = result8[['CustomerID','Invoices']].sort_values('Invoices',ascending = False).head(10)
print(result11)
```
