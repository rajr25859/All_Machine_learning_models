**General Life-cycle of an experimental Supervised ML Project**
- Data Import
- Handle Missing Data
- Non-Numeric and Categorical Data Management 
- Standardization
- Split Data into Features (Independent Variables/Predictors/X) and Labels (Dependent Variable/y)
- Divide X, y into training and testing part
- Build Model
- Train the Model with Training Data
- Predict the Result on Testing Features
- Evaluate the Model by comparing predicted labels with actual (testing) labels

### Data Sources
- Flat Files (txt, csv, tsv etc.): Pandas
- Spreadsheets (Excel, ODS): Pandas
- RDBMS: DB API, SQLAlchemy, Pandas
- NoSQL -> MongoDB, Firebase, CouchDB, Redis etc
- HDFS -> Pandas
- Pickled Data -> pickle, Pandas
- JSON, XML or API -> json, xml, Pandas
- Unstructured Data: Web (requests, BeautifulSoup, Scrapy), File System (os)
