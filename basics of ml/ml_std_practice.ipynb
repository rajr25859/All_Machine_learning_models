```python
import numpy as np
import pandas as pd
```

```python
a = np.array([12,15,13,14,15,16])
b = np.array([1,2,1,3,-1,0])
c = np.array([1000,1035,900,790,815,1024])
```

```python
data = np.array([a,b,c]).T
data
```

```python
a,b,c
```

```python
mean_a = np.mean(a)
mean_b = np.mean(b)
mean_c = np.mean(c)
```

```python
std_a = np.std(a)
std_b = np.std(b)
std_c = np.std(c)
```

```python
score_a = (a-mean_a)/std_a
score_b = (b-mean_b)/std_b
score_c = (c-mean_c)/std_c
```

```python
score_a
```

```python
score_c
```
